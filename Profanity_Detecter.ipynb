```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForTokenClassification

tokenizer = AutoTokenizer.from_pretrained("troesy/toxicbert-hatexplain-label-all-tokens-False-3epoch")
model = AutoModelForTokenClassification.from_pretrained("troesy/toxicbert-hatexplain-label-all-tokens-False-3epoch")
```

```python
input_str = "“@skimaskjojo: “@1waynostra: Dat bird bitch”HEAVY BIRDCAGE BITCH” byyyyyyrrrrrrrrrrrrr bitch fuck you🐥"
```

```python
tokens = tokenizer(input_str, return_tensors="pt")
output = model(**tokens)
output
```

```python
output.logits.shape
```

```python
tokens = tokenizer(input_str, return_tensors="tf")
decoded_ouput = tokenizer.decode(tokens.input_ids[0])
```

```python
tokens.input_ids
```

```python
import torch
var = torch.tensor(output.logits)
probs = torch.nn.functional.softmax(var, dim=-1)
prediction = torch.argmax(probs, dim=-1)[0].tolist()
```

```python
len(decoded_ouput)
```

```python
input_str.split(" ")
```

```python
for label, word in zip(prediction, tokens.input_ids[0]):
  if label!=0:
    print(tokenizer.decode(word))
```
